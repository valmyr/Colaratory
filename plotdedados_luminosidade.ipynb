```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('/content/dados.txt','r')
lumi1 = []
lumi2 = []
lumi3 = []
lumi4 = []
lumimed = []

for lin in file:
  lumi1.append((lin.split(',')[7]))
  lumi2.append((lin.split(',')[8]))
  lumi3.append((lin.split(',')[9]))
  lumi4.append((lin.split(',')[10]))
  lumimed.append((lin.split(',')[11]))
t = np.arange(0,3.6,0.1)
```

```python
from scipy.interpolate import interp1d
import pylab as pl
def l(t,lumi1,s):
  measured_time = t
  noise = (np.random.random(10)*2 - 1) * 1e-1
  measures =lumi1
  linear_interp = interp1d(measured_time, measures)
  computed_time = np.linspace(0, 1, 50)
  cubic_interp = interp1d(measured_time, measures, kind='cubic')
  cubic_results = cubic_interp(computed_time)
  pl.plot(computed_time, cubic_results, label=s)
  pl.legend()
```

```python
l(t,lumi1,'lumisonidade sensor1')
l(t,lumi2,'lumisonidade sensor2')
l(t,lumi3,'lumisonidade sensor3')
l(t,lumi4,'lumisonidade sensor4')
l(t,lumi4,'lumisonidade m√©dia')
```
