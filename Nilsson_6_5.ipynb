```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Intervalos em análise
t0 = np.arange(start = 0,stop = 2.1,step= 0.1,dtype = float)
t1 = np.arange(start = 2,stop = 6.1,step= 0.1,dtype = float)
t2 = np.arange(start = 6,stop = 10.1,step= 0.1,dtype = float)
t3 = np.arange(start = 10,stop = 12.1,step= 0.1,dtype = float)
```

```python
#Para a tensão VL(t)
VL_0 = lambda t:-25*t
VL_1 = lambda t:25*t-100
VL_2 = lambda t: -25*t+200
VL_3 = lambda t:25*t-12*25
VL_4 = lambda t:0
```

```python
#Para a corrente iL(t)
iL_0 = lambda t:-25*t**2
iL_1 = lambda t:31.25*t**2-250*t+275
iL_2 = lambda t:-31.25*t**2+500*t-1975
iL_3 = lambda t:31.25*t**2-750*t+4275
iL_4 = lambda t:0
```

```python
plt.subplot(4,1,1)
plt.plot(t0,VL_0(t0))
plt.plot(t1,VL_1(t1))
plt.plot(t2,VL_2(t2))
plt.plot(t3,VL_3(t3))
plt.xlabel("t [ms]")
plt.ylabel("VL(t) [V]")
plt.subplot(4,1,2)
plt.xlabel("t [ms]")
plt.ylabel("iL(t) [A]")
plt.plot(t0,iL_0(t0))
plt.plot(t1,iL_1(t1))
plt.plot(t2,iL_2(t2))
plt.plot(t3,iL_3(t3))
plt.subplot(4,1,3)
plt.xlabel("t [ms]")
plt.ylabel("P(t) [W]")
plt.plot(t0,iL_0(t0)*VL_0(t0))
plt.plot(t1,iL_1(t1)*VL_1(t1))
plt.plot(t2,iL_2(t2)*VL_2(t2))
plt.plot(t3,iL_3(t3)*(t3))
plt.subplot(4,1,4)
plt.xlabel("t [ms]")
plt.ylabel("W(t) [J]")
plt.plot(t0,(2.5*iL_0(t0)**2)/2)
plt.plot(t1,(2.5*iL_1(t1)**2)/2)
plt.plot(t2,(2.5*iL_2(t2)**2)/2)
plt.plot(t3,(2.5*iL_3(t3)**2)/2)
plt.show()
```
